```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating a plot
ax = df.plot(x="Molecular Weight", y="measured log solubility in mols per litre", label="measured", kind="scatter", color="purple")
df.plot(x="Molecular Weight", y="ESOL predicted log solubility in mols per litre", label="pridicted", kind="scatter", ax=ax, color="black")

# Calculating and ploting regression lines
x = df["Molecular Weight"]
y1 = df["measured log solubility in mols per litre"]
y2 = df["ESOL predicted log solubility in mols per litre"]

m1, b1 = np.polyfit(x, y1, 1)
m2, b2 = np.polyfit(x, y2, 1)

plt.plot(x, m1 * x + b1, color='pink', label="measured log solubility in mols per litre (measured)")
plt.plot(x, m2 * x + b2, color='grey', label="ESOL predicted log solubility in mols per litre(pridicted)")

# Customizing the plot
plt.title("Two Curves with Regression Lines")
plt.xlabel("Molecular Weight")
plt.ylabel("Solubility")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Selecting the column for the histogram
column_name = "ESOL predicted log solubility in mols per litre"

# Create a figure and axis
fig, ax = plt.subplots()

# Create a histogram of the selected column
ax.hist(df[column_name], bins=20, edgecolor='black', alpha=0.7, label='Histogram')

# Calculating mean, median, and standard deviation
mean_value = df[column_name].mean()
median_value = df[column_name].median()
std_value = df[column_name].std()

# Ploting vertical lines for mean, median, and standard deviation
ax.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Mean')
ax.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label='Median')
ax.axvline(std_value, color='blue', linestyle='dashed', linewidth=2, label='StdDev')

# Calculating the mode
mode_value = df[column_name].mode().iloc[0]
ax.axvline(mode_value, color='purple', linestyle='dashed', linewidth=2, label='Mode')

# Customizing the plot
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Histogram with Mean, Median, StdDev, and Mode for {column_name}')
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load your data into a DataFrame
# Assuming you have a DataFrame called 'df' with the specified column types

# Select the numeric columns for clustering
numeric_columns = df.iloc[:, 1:7]

# Standardize the numeric data
scaler = StandardScaler()
numeric_columns_scaled = scaler.fit_transform(numeric_columns)

# Perform K-Means clustering
kmeans = KMeans(n_clusters=3)  # Adjust the number of clusters as needed
clusters = kmeans.fit_predict(numeric_columns_scaled)

# Create a scatter plot to visualize clusters
plt.scatter(numeric_columns_scaled[:, 0], numeric_columns_scaled[:, 1], c=clusters, cmap='rainbow')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Cluster Analysis')

plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load your data into a DataFrame
# Assuming you have a DataFrame called 'df' with the specified column types

# Select the numeric columns for PCA
numeric_columns = df.iloc[:, 1:6]  # Select columns 2 to 6

# Standardize the numeric data
scaler = StandardScaler()
numeric_columns_scaled = scaler.fit_transform(numeric_columns)

# Perform PCA
pca = PCA(n_components=2)  # Reduce to 2 principal components for visualization
principal_components = pca.fit_transform(numeric_columns_scaled)

# Create a scatter plot to visualize the data in the reduced space
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Analysis')

plt.show()

# Variance explained by each principal component
explained_variance = pca.explained_variance_ratio_
print(f'Variance explained by PC1: {explained_variance[0]:.2f}')
print(f'Variance explained by PC2: {explained_variance[1]:.2f}')
```
